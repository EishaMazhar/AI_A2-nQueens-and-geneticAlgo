```python
# helper fuction to print all the solutions and will return the no. of solutions to display

def printAllSolutions(myBoardObj):

    noOfQueens = myBoardObj.getSize()
 
    # baseCase
    if noOfQueens == myBoardObj.QueensCount():
        myBoardObj.display()
        print("\n")
        return 1
 
    solutions = 0
    # place queen in next row
    for i in range(noOfQueens):
        
          #check if it's safely placed in column
        if myBoardObj.isSafe(i):
            
            myBoardObj.MoveToNextRow(i)
            #recursive call
            solutions += printAllSolutions(myBoardObj) 
            myBoardObj.RemoveFromCurrentRow()
 
    return solutions
```

```python
def printAllPossibleSolutions(numberOfQueens):
    
     #iniltializing the class
    ObjBoard = nQueenBoard(numberOfQueens)
    
    # get total no of solutions after printing them in printAllSolutions(objBoard)
    TotalSolutions = printAllSolutions(ObjBoard)
    
    print('Number of possible solutions -->', TotalSolutions)
 
```

```python
# main 

nQueens = int(input('Enter the number of queens : '))
print()
printAllPossibleSolutions(nQueens)
```
